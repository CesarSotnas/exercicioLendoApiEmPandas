####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
# 1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```),
# e que devolve os dados da consulta no formato *DataFrame*.

#1) Seu código aqui
def consulta_api(estado, tipo_interesse):
    
    if tipo_interesse.lower() == 'estado':
         url = f'http://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={estado.upper()}&tipo_interessado=Estado'

    elif tipo_interesse.lower() == 'município':
         url = f'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={estado.upper()}&tipo_interessado=Munic%C3%ADpio'

    r = requests.get(url)
    dados_json = r.json()
    
    return dados_json['items']  
````

````python
df = pd.DataFrame(consulta_api('mG', 'EstAdo'))
````

````python
df.head()
````

````python
# 2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 
# 'Arquivado por decurso de prazo' estão registradas?

# 2) Seu código aqui
df['status'].value_counts()
````

````python
# 3. Qual é o município da Bahia com mais solicitações deferidas?
# 3) Seu código aqui

df = pd.DataFrame(consulta_api('BA', 'município'))
````

````python
df.head()
````

````python
df_filtro = df.loc[df['status'] == 'Deferido']
````

````python
print('Bairros com mais status deferido: ')
df_filtro[['interessado', 'status']].value_counts()
````

````python
# 4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'
# 4) Seu código aqui
df = pd.DataFrame(consulta_api('BA', 'estado'))
````
